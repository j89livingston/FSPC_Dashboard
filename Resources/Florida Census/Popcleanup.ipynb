# Florida Census data cleanup for FSPC Analysis

We need track the total number of residents per county per year for our analysis. Currently, we are only tracking the last 10 years. This data has incomplete stats for 2020 but are working with 2019 and below.

```python
#import dependencies
import pandas as pd
import numpy as np
```

```python
#import data
data = pd.read_excel("FLcopops.xls",None, usecols="A:B")
```

```python
# Check multiple sheet names
data.keys()
```

```python
# select sheets for data merge
c20 = data['2020 Revised']
c19 = data['2019']
c18 = data['2018']
c17 = data['2017']
c16 = data['2016']
c15 = data['2015']
c14 = data['2014']
c13 = data['2013']
c12 = data['2012']
c11 = data['2011']
```

```python
# Rename Column names, the lazy way
c20.columns = ['County', '2020']
c19.columns = ['County', '2019']
c18.columns = ['County', '2018']
c17.columns = ['County', '2017']
c16.columns = ['County', '2016']
c15.columns = ['County', '2015']
c14.columns = ['County', '2014']
c13.columns = ['County', '2013']
c12.columns = ['County', '2012']
c11.columns = ['County', '2011']
```

```python
#Check it
print(c19.head())
print(c19.tail(7))
```

```python
# merge dataframes
df = pd.concat([c19,c20,c18,c17,c16,c15,c14,c13,c12,c11], True)
df.head()
```

```python
#Remove duplicated columns
df = df.loc[:,~df.columns.duplicated()]
df
```

```python
#Shift dataframe up 2 to remove old names
df = df.shift(-2)
df
```

```python
# Check the bottom of dataframe to see if anything needs to be removed
df.tail(10)
```

```python
# group dataframe by usable data
df = df.loc[:67]
df
```

```python
df = df[['County', '2020','2019','2018','2017','2016','2015','2014','2013','2012','2011']]
df
```

```python
df
```

```python
#Unpivot the columns to create a Years column
pivdata = pd.melt(df, 
            id_vars='County', 
            value_vars=list(df.columns[1:]), # list of days of the week
            var_name='Years', 
            value_name='Population')
```

```python
pivdata.head(10)
```

```python
pivdata.count()
```

```python
# Save file
pivdata.to_csv(r'Popcleanup.csv', encoding='utf-8', index=False)
```
