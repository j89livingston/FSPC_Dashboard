```python
# Dependecies and Setup
import pandas as pd
import numpy as np

# File to load
data = "Resident Deaths by Residence County by Year.xlsx"

# Read the data into a Pandas DataFrame
data_df = pd.read_excel(data)

# Check data
data_df.head(10)
```

```python
# Drop last row. Data can generate the total in other ways so we wont need it.
data_df = data_df.drop('Unnamed: 11', 1)
data_df.head(5)
```

```python
#Remove useless top rows
data_df = data_df.shift(-3)
data_df.head(5)
```

```python
#Remove useless bottom rows
data_df = data_df.loc[:67]
data_df.tail(10)
```

```python
# Rename Columns
data_df.columns = ['County','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020']
data_df.head(5)
```

```python
#Unpivot the columns to create a Years column
pivdata = pd.melt(data_df, 
            id_vars='County', 
            value_vars=list(data_df.columns[1:]), 
            var_name='Years', 
            value_name='Death Count')
```

```python
# Check it
pivdata.head()
```

```python
# Count it 
pivdata.count()
```

```python
pivdata.to_csv(r'./Fldeathbycounty.csv', encoding='utf-8', index=False)
```
