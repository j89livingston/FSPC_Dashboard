```python
# Dependecies and Setup
import pandas as pd
import numpy as np

# File to load
survey_data = "Middle School Survey all years.csv"

# Read the data into a Pandas DataFrame
middle_data_df = pd.read_csv(survey_data)

# Check data
middle_data_df.head(10)
```

# Clean Data

```python
print(middle_data_df.dtypes)
```

```python
#Display data info
middle_data_df.info()
```

```python
#Select columns to create a new dataframe
new_df=middle_data_df[['year', 'record', 'age', 'grade', 'qn14', 'qn15', 'qn16']]

print(new_df)
```

```python
#Rename the columns
middle_df= new_df.rename(columns={'qn14': 'consideration', 'qn15': 'plan', 'qn16': 'attempt'})

middle_df
```

```python
# Check for null values
middle_df.isnull().sum()
```

```python
#Create a dataframe without nulls
new_middle_df= middle_df.dropna()
new_middle_df
```

```python
#Export new dataframe to CCSV file
new_middle_df.to_csv("Clean Attempts Middle School")
```
