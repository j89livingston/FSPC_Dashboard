```python
# Dependecies and Setup
import pandas as pd
import numpy as np

# File to load
data = "suicidebyfirearm.csv"

# Read the data into a Pandas DataFrame
data_df = pd.read_csv(data)

# Check data
data_df.head(10)
```

# Clean Data

```python
# rename columns
new_data_df= data_df.rename(columns={'Age-adjusted Deaths from Suicide by Firearms Discharge':'county',
                                           '1':'2020 Count', '2':'2020 Rate', '3':'2019 Count', 
                                           '4':'2019 Rate', '5':'2018 Count', '6':'2018 Rate', 
                        '7':'2017 Count', '8':'2017 Rate', '9':'2016 Count', '10':'2016 Rate', '11':'2015 Count', 
                        '12':'2015 Rate', '13':'2014 Count', '14':'2014 Rate', '15':'2013 Count', '16':'2013 Rate', 
                        '17':'2012 Count', '18':'2012 Rate', '19':'2011 Count', '20':'2011 Rate' })

new_data_df
```

```python
firearm_data= new_data_df.drop([0,1,2])
firearm_data
```

```python
# Check for null values
firearm_data.isnull().sum()
```

```python
#Export new dataframe to CCSV file
#firearm_data.to_csv("Clean suicide by firearm.csv")
```

# Create new dataframe pivdatarate

```python
firearm_data.columns = ['Placeholder', 
                '2020 Count', '2020 Rate', 
                '2019 Count', '2019 Rate',
                '2018 Count', '2018 Rate',
                '2017 Count', '2017 Rate',
                '2016 Count', '2016 Rate',
                '2015 Count', '2015 Rate',
                '2014 Count', '2014 Rate',
                '2013 Count', '2013 Rate',
                '2012 Count', '2012 Rate',
                '2011 Count', '2011 Rate',]
```

```python
# Drop all rows containing Count
onlyrate = firearm_data[firearm_data.columns.drop(list(firearm_data.filter(regex='Count')))]
```

```python
#Rename columns with only years.
onlyrate.columns = ['County', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', 
                    '2012', '2011']
onlyrate.columns
```

```python
#Unpivot the columns to create a Years column
pivdatarate = pd.melt(onlyrate, 
            id_vars='County', 
            value_vars=list(onlyrate.columns[1:]), 
            var_name='Years', 
            value_name='Rate')
  
pivdatarate   
```

```python
#Check it
pivdatarate.count()
```

# Create new dataframe pivdatacount

```python
# Drop all rows containing Rate
onlycount = firearm_data[firearm_data.columns.drop(list(firearm_data.filter(regex='Rate')))]
```

```python
#Rename columns with only years.
onlycount.columns = ['County', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', 
                    '2012', '2011']
onlycount.columns
```

```python
#Unpivot the columns to create a Years column
pivdatacount = pd.melt(onlycount, 
            id_vars='County', 
            value_vars=list(onlycount.columns[1:]), # list of days of the week
            var_name='Years', 
            value_name='Count')
  
pivdatacount     
```

# Merge to create one dataframe

```python
#Merge pivdatarate with pivdatacount to make one dataframe
merge_df = pd.merge(pivdatarate, pivdatacount)
merge_df
```

```python
merge_df['County'] = merge_df['County'].str.replace('Desoto', 'DeSoto')
```

```python
#Export new dataframe to CCSV file
merge_df.to_csv("Clean suicide by firearm.csv")
```
