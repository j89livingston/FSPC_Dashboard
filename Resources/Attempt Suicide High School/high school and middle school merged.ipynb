```python
# Dependecies and Setup
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# File to load
high_survey_data = "Clean Attempts High School.csv"
middle_survey_data = "Clean Attempts Middle School.csv"

# Read the High School and Middle Data and store into a Pandas DataFrame
high_data_df = pd.read_csv(high_survey_data)
middle_data_df = pd.read_csv(middle_survey_data)
```

```python
# Combine the data into a single dataset.
both_df = pd.merge(high_data_df, middle_data_df, on=["year", "year"])
both_df
```

```python
#Export new dataframe to CSV file
both_df.to_csv("Clean Attempts Both School.csv", encoding='utf-8', index=False)
```
