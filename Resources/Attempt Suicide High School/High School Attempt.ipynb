```python
# Dependecies and Setup
import pandas as pd
import numpy as np

# File to load
survey_data = "High School Survey all years.csv"

# Read the data into a Pandas DataFrame
high_data_df = pd.read_csv(survey_data)

# Check data
high_data_df.head(10)
```

```python
print(high_data_df.dtypes)
```

```python
#Display data info
high_data_df.info()
```

```python
#Select columns to create a new dataframe
new_df=high_data_df[['year', 'record', 'age', 'grade', 'qn26', 'qn27', 'qn28']]

print(new_df)
```

```python
#Rename the columns (1=yes, 2=No)
high_df= new_df.rename(columns={'qn26': 'consideration', 'qn27': 'plan', 'qn28': 'attempt'})

high_df
```

```python
# Check for null values
high_df.isnull().sum()
```

```python
#Create a dataframe without nulls
new_high_df= high_df.dropna()
new_high_df
```
