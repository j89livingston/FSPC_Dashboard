```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create DataFrame
suicideData = "suicide death cost per county.csv"
suicide_df = pd.read_csv(suicideData)

suicide_df
```

```python
#Explore Data
print(suicide_df.dtypes)
```

```python
#Convert Strings to Numeric
suicide_df['Population'] = pd.to_numeric(suicide_df['Population'],errors="coerce").fillna(0)
suicide_df['Deaths'] = pd.to_numeric(suicide_df['Deaths'],errors="coerce").fillna(0)
suicide_df['Med_Cost'] = pd.to_numeric(suicide_df['Med_Cost'],errors="coerce").fillna(0)
suicide_df['Work_Cost'] = pd.to_numeric(suicide_df['Work_Cost'],errors="coerce").fillna(0)

#Check Conversion
print(suicide_df.dtypes)
```

```python
#Remove Header Info (First 21 Rows)
suicide_df = suicide_df.iloc[21: , :]
suicide_df
```

```python
#Deaths by County
suicide_df.plot(x="County", y="Deaths", kind="bar", figsize=(20,15))
plt.show()
```

```python
#Deaths by Population
suicide_df.plot(x="Population", y="Deaths", kind="bar", figsize=(20,15))
plt.show()
```

```python
#Med Costs by County
suicide_df.plot(x="County", y="Med_Cost", kind="bar", figsize=(20,15))
plt.show()
```

```python
#Work Loss Costs by County
suicide_df.plot(x="County", y="Work_Cost", kind="bar", figsize=(20,15))
plt.show()
```

```python
#All Costs by County
suicide_df.plot(x="County", y=["Med_Cost","Work_Cost"], kind="bar", figsize=(20,15))
plt.show()
```

```python
#Deaths per County & Population By County
suicide_df.plot(x="County", y=["Population","Deaths"], kind="bar", figsize=(20,15))
plt.show()
```
