# Count and Rate Cleanup

```python
#import dependencies
import pandas as pd
```

```python
# import data
data = pd.read_excel('./floridasuicidecountandrate.xlsx')
```

```python
#Check it
data.head()
```

```python
# Rename Columns for easier cleanup
data.columns = ['County', 
                '2020 Count', '2020 Rate', 
                '2019 Count', '2019 Rate',
                '2018 Count', '2018 Rate',
                '2017 Count', '2017 Rate',
                '2016 Count', '2016 Rate',
                '2015 Count', '2015 Rate',
                '2014 Count', '2014 Rate',
                '2013 Count', '2013 Rate',
                '2012 Count', '2012 Rate',
                '2011 Count', '2011 Rate',]
```

```python
#Shift up two rows to get rid of useless rows
data = data.shift(-2)
```

```python
#Check it again
data.head()
```

```python
data['County'] = data['County'].str.replace('Saint Johns', 'St. Johns')
data['County'] = data['County'].str.replace('Saint Lucie', 'St. Lucie')
data['County'] = data['County'].str.replace('Desoto', 'DeSoto')
```

```python
print(list(data['County']))
```

# Create new dataframe only containing count

```python
onlycount = data.copy()
```

```python
# Drop all rows containing rate
onlycount = onlycount[onlycount.columns.drop(list(data.filter(regex='Rate')))]
```

```python
#Check it
onlycount.head()
```

```python
#Rename columns with only years.
onlycount.columns = ['County', '2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011']
```

```python
#Check it
onlycount.head()
```

```python
#Unpivot the columns to create a Years column
pivdata = pd.melt(onlycount, 
            id_vars='County', 
            value_vars=list(onlycount.columns[1:]), # list of days of the week
            var_name='Years', 
            value_name='Count')
```

```python
#Drop NA
pivdata = pivdata.dropna()
```

```python
#Check it
pivdata.count()
```

```python
pivdata.to_csv(r'./Flsuicidecountonly.csv', encoding='utf-8', index=False)
```
